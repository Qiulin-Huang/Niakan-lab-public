```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
Annotated_peaks = pd.read_csv('n2_n3_overlapping_annotated_hg19gtf.tsv', sep = '\t', comment = "#", index_col = 0)
```

```python
Annotated_peaks
```

```python
Promoter_binding_list = Annotated_peaks[['Annotation', 'Gene Name', 'Peak Score']]
```

```python
Promoter_binding_list
```

```python
Promoter_binding_list = Promoter_binding_list[Promoter_binding_list['Annotation'].str.contains('promoter|TSS', case=False, na=False)] # Get all the data that contains promoter or TSS in the Annotation column
```

```python
Promoter_binding_list
```

```python
Promoter_binding_list.to_csv('Peaks_at_promoters.tsv', sep = '\t', index = True)
```

```python
Peak_based_gene_rank = Promoter_binding_list[['Gene Name', 'Peak Score']]
```

```python
Peak_based_gene_rank
```

```python
Peak_based_gene_rank = Peak_based_gene_rank[~Peak_based_gene_rank['Gene Name'].str.contains('ENSG', case=False, na=False)] # Get all the data that does not contain Ensembl id
```

```python
Peak_based_gene_rank
```

```python
Peak_based_gene_rank.to_csv('Peak_based_gene_rank.tsv', sep = '\t', index = False)
```

```python
Peak_based_gene_rank = pd.read_csv('Peak_based_gene_rank.tsv', sep = '\t', index_col = False)
Peak_based_gene_rank
```

```python
Peak_based_gene_rank_top100 = Peak_based_gene_rank[0:100]
Peak_based_gene_rank_top100
```

```python
Peak_based_gene_rank.index = Peak_based_gene_rank['Gene Name']
```

```python

Peak_based_gene_rank[4:55].plot(kind='bar', figsize=(20, 10), fontsize=20)

plt.ylabel('Normalized peak intensity', fontsize=20)
plt.title(' ')

plt.tight_layout()

plt.savefig('/Bioinformatics_D/Figures/SALL2_binding_genes1.pdf', bbox_inches='tight', dpi=300)
plt.savefig('/Bioinformatics_D/Figures/SALL2_binding_genes1.png', bbox_inches='tight', dpi=300)
plt.show()
```
