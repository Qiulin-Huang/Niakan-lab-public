```python
import gseapy as gp
import pandas as pd
import numpy as np
```

```python
# Load the DE analysis result from DESeq2
de_results = pd.read_csv('Day_10_DEG_df.csv')
de_results
```

```python
de_results = de_results.dropna(subset=['log2FoldChange']) # Remove rows that have NA in lop2FoldChange
de_results = de_results.sort_values(by='log2FoldChange', ascending=False) # Sort the dataframe based the log2FoldChange
gene_list = de_results[['GeneID', 'log2FoldChange']] # Extract GeneID and log2FoldChange into a new dataframe
gene_list = gene_list.set_index('GeneID') # Set the index into GeneID
gene_list = gene_list['log2FoldChange'] # Get the indexed list of log2FoldChange
```

```python
gene_list
```

```python
# Run GSEA Preranked

result_path = 'GSEA_analysis_day_10_GO'

pre_res = gp.prerank(
    rnk=gene_list,  # Ranked gene list
    gene_sets='c5.all.v2024.1.Hs.symbols.gmt',  # Gene set database
    outdir=result_path,  # Output directory
    permutation_num=1000,  # Number of permutations
    format='png',  # Output format of figures
    seed=123  # Random seed for reproducibility
)
```

```python
prerank_report = pd.read_csv(result_path + '/gseapy.gene_set.prerank.report.csv')
prerank_report
```

```python
# This is the same as the prerank_report
gsea_results = pre_res.res2d
gsea_results
```

```python
gsea_results_BP = gsea_results[gsea_results['Term'].str.contains('GOBP', na = False)]
gsea_results_MF = gsea_results[gsea_results['Term'].str.contains('GOMF', na = False)]
gsea_results_HP = gsea_results[gsea_results['Term'].str.contains('HP_', na = False)]
gsea_results_CC = gsea_results[gsea_results['Term'].str.contains('GOCC', na = False)]
```

```python
gsea_results_rankby_NES = gsea_results.sort_values(by='NES', ascending=False) # Sort base on NES score
gsea_results_rankby_NES
```

```python
gsea_results_BP_rankby_NES = gsea_results_BP.sort_values(by='NES', ascending=False) # Sort base on NES score
gsea_results_MF_rankby_NES = gsea_results_MF.sort_values(by='NES', ascending=False) # Sort base on NES score
gsea_results_CC_rankby_NES = gsea_results_CC.sort_values(by='NES', ascending=False) # Sort base on NES score
```

```python
pre_res.plot(terms = gsea_results_rankby_NES['Term'][0:10], ofname='gsea_results/Higest_10_NES_enrichment_plot.png')
pre_res.plot(terms = gsea_results_rankby_NES['Term'][len(gsea_results)-10:len(gsea_results)], ofname='gsea_results/Lowest_10_NES_enrichment_plot.png')
pre_res.plot(terms = gsea_results['Term'][0:10], ofname='gsea_results/Higest_10_FWERpval_enrichment_plot.png')
pre_res.plot(terms = gsea_results['Term'][len(gsea_results)-10:len(gsea_results)], ofname='gsea_results/Lowest_10_FWERpval_enrichment_plot.png')
```
