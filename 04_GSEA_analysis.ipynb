```python
import gseapy as gp
import pandas as pd
import numpy as np
```

```python
# Load the DE analysis result from DESeq2
de_results = pd.read_csv('10X_DEG_SALL2KO.csv', index_col = 0)
de_results
```

```python
de_results = de_results.dropna(subset=['logfoldchanges']) # Remove rows that have NA in lop2FoldChange
de_results = de_results.sort_values(by='logfoldchanges', ascending=False) # Sort the dataframe based the log2FoldChange
gene_list = de_results[['names', 'logfoldchanges']] # Extract GeneID and log2FoldChange into a new dataframe
gene_list = gene_list.set_index('names') # Set the index into GeneID
gene_list = gene_list['logfoldchanges'] # Get the indexed list of log2FoldChange
```

```python
gene_list
```

```python
# Run GSEA Preranked

result_path = 'GSEA_analysis_SALL2KO_Pathways'

pre_res = gp.prerank(
    rnk=gene_list,  # Ranked gene list
    gene_sets='c2.all.v2025.1.Hs.symbols.gmt',  # Gene set database
    outdir=result_path,  # Output directory
    permutation_num=1000,  # Number of permutations
    format='png',  # Output format of figures
    seed=123  # Random seed for reproducibility
)
```

```python
# This is the same as the prerank_report
gsea_results = pre_res.res2d
gsea_results
```

```python
gsea_results_rankby_NES = gsea_results.sort_values(by='NES', ascending=False) # Sort base on NES score
gsea_results_rankby_NES
```

```python
pre_res.plot(terms = gsea_results_rankby_NES['Term'][0:10], ofname=result_path + '/gsea_results/Higest_10_NES_enrichment_plot.pdf')
```

```python
gsea_results_rankby_NES_qval = gsea_results_rankby_NES[gsea_results_rankby_NES['FDR q-val'] < 0.25] # 0.25 is considered as the significant threshold for FDR q-val
gsea_results_rankby_NES_qval
```

```python
pre_res.plot(terms = gsea_results_rankby_NES_qval['Term'][0:10], ofname= result_path + '/gsea_results/qval_sig_up_enrichment_plot.pdf') # GSEA plots for significanly upregulated pathways
pre_res.plot(terms = gsea_results_rankby_NES_qval['Term'][len(gsea_results_rankby_NES_qval)-10:len(gsea_results_rankby_NES_qval)], ofname= result_path + '/gsea_results/qval_sig_down_enrichment_plot.pdf') # GSEA plots for significanly downregulated pathways
```
